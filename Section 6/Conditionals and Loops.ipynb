```python
legs = int(input("Enter the total number of legs: "))
animals = int(input("Enter the total number of animals: "))

pigs = legs // 4
chickens = animals - pigs

while legs != 4 * pigs + 2 * chickens:
    pigs -= 1
    chickens += 1
    if pigs < 0:
        raise ValueError("There is no solution for these inputs.")
print("\nNumber of chickens = ", chickens)
print("\nNumber of pigs = ", pigs)
print("\nNumber of legs = ", 4 * pigs + 2 * chickens)
print("\nNumber of animals = ", pigs + chickens)
```

```python
while True:
    try:
        legs = int(input("Enter the total number of legs: "))
        assert legs % 2 == 0
        break
    except AssertionError:
        print("Total number of legs must be an even number")
while True:
    try:
        animals = int(input("Enter the total number of animals: "))
        assert 4 * animals >= legs and 2 * animals <= legs
        break
    except AssertionError:
        print("Number of animals must be >= {} and <= {}".format((legs + 2) // 4, legs // 2))

pigs = legs // 4  # Maximum possible number of pigs
chickens = animals - pigs

while legs != 4*pigs + 2*chickens:
    pigs -= 1
    chickens += 1
    if pigs < 0:
        raise ValueError("There is no solution for these inputs")  # It helps you debug

print("\nNumber of chickens is: ", chickens)
print("Number of pigs is: ", pigs)
print("Number of legs is: ", 4 * pigs + 2 * chickens)
print("Number of animals is: ", pigs + chickens)
```
