```python
legs = int(input("Enter the total number of legs: "))
animals = int(input("Enter the total number of animals: "))

pigs = legs // 4
chickens = animals - pigs

while legs != 4 * pigs + 2 * chickens:
    pigs -= 1
    chickens += 1
    if pigs < 0:
        raise ValueError("There is no solution for these inputs.")
print("\nNumber of chickens = ", chickens)
print("\nNumber of pigs = ", pigs)
print("\nNumber of legs = ", 4 * pigs + 2 * chickens)
print("\nNumber of animals = ", pigs + chickens)
```

```python
import random
# Make a list of k integers radomly chosen from range
nums = random.choices(range(-1000, 1000), k=10)
print(nums)

max = nums[0]
for n in nums:
    if n > max:
        max = n
print("The maximum value is:", max)
```
