```python
while True:
    try:
        n = int(input("Please enter an integer positive number: "))
        assert n >= 0  # Include the condition on the type
        break
    except (AssertionError, ValueError):
        print("Try again")

f = 1

for i in range(n, 0, -1):
    f = f*i
print(f"The factorial of {n} is {f}")
```

```python
n = int(input("Input an integer > 1: "))  # Include conditions on this n
i = 2

while (n % i) != 0:
    i += 1
    
if n == i:
    print(f"{n} is a prime number")
else:
    print(f"The smallest prime factor of {n} is {i}")
```

```python
'''Compound interest'''

P = float(input("Please enter the initial investment: "))
r = float(input("\nPlease enter the interest rate (%): "))
n = int(input("\nPlease enter the number of years the money will be invested: "))
print(P, r, n)

#r = r / 100

for t in range(0, n + 1):
    A = P * ((1 + r/n)**(n*t))
    print(f"The earnings after {t} years are {A:.2f}")
    if t == 10:
        print(A, P, r, n, t)
```

```python
'''Compound interest'''

P = float(input("Please enter the initial investment: "))
r = float(input("\nPlease enter the interest rate (%): "))
n = int(input("\nPlease enter the number of years the money will be invested: "))

r = r / 100
print("Anually($)" + '         ' + "Monthly($)" + '         ' + "Daily($)")
for t in range(0, n + 1):
    A_a = P * ((1 + r/1)**(1*t))
    A_m = P * ((1 + r/12)**(12*t))
    A_d = P * ((1 + r/365)**(365*t))
    print(f"\n{A_a:8.2f} {A_m:19.2f} {A_d:17.2f}")
```

```python
'''Highest and Lowest in dictionary'''

ps = {"Adele": 35, "Elton": 76, "Taylor": 34, "Billie": 22} # What if there are two persons with the same age?

mx = list(ps.values())[0]
mn = list(ps.values())[0]
name_max = list(ps.keys())[0]
name_min = list(ps.keys())[0]

for n,i in zip(ps.keys(), ps.values()):

    print(ps[n])
    if ps[n] > mx:
        mx = ps[n]
        name_max = n
    
    if ps[n] < mn:
        mn = ps[n]
        name_min = n
        
print(f"The oldest person is {name_max} and the youngest person is {name_min}")
```
