```python
money = float(input("How mouch money (in dollars) in your lunch account?"))
day = int(input("What day of the month is today?"))
days = int(input("How many days in this month"))
spend = money/(days-day+1)
print(f"You can spend ${spend:0.2f} each day for the rest of the month")
```

```python
money = float(input("How mouch money (in dollars) in your lunch account?"))
day = int(input("What day of the month is today?"))
month = input("What month is it today?")

months = {"January": 31, "February": 28, "March": 31, "April": 30, "May": 31,
          "June": 30, "July": 31, "August": 31, "September": 30, "October": 31,
          "November": 30, "December": 31}

spend = money/(months[month]-day+1)
print(f"You can spend ${spend:0.2f} each day for the rest of the month")
```
