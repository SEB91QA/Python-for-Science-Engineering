```python
""" Calculates time , electrical energy used , and cost of electricity
for a trip in an electric vehicle"""
distance = float(input("Input trip distance (miles): "))

mpk = 3.9  # [miles/kilowatt -h] car mileage
speed = 60.  # [miles/h] average speed
cost_per_kWh = 0.22  # [$/kW-h] price of electricity
time = distance / speed  # [hours]
energy = distance / mpk  # [kW-h]
cost = energy * cost_per_kWh  # [$]

print("\nDuration of trip = {0:0.1f} hours".format(time))  # Formatting
print("Electricity used = {0:0.1f} kW-h (@ {1:0.2f} miles/kW-h)".format(energy, mpk))
print("Cost of electricity = ${0:0.2f} (@ ${1:0.2f}/kW-h)".format(cost, cost_per_kWh))
```

```python
import numpy as np

dataPt, time, height, error = np.loadtxt("mydata.txt", skiprows=5, unpack=True)
np.savetxt("mydataout.txt", list(zip(dataPt, time, height, error)), fmt="%12d %10.2f %12.0f %12.1f")
```

```python
dataPt, time, height, error = np.loadtxt("mydata.txt", skiprows=5, unpack=True)

info = 'Data for falling mass experiment'
info += '\nDate: 04-Aug-2021'
info += '\nData taken by SEB'
info += '\n\n data point time (sec) height (mm)'
info += 'uncertainty (mm)'

np.savetxt('read_write_my_data_header.txt', 
           list(zip(dataPt, time, height, error)),
           header=info, fmt="%12.1f")
```
