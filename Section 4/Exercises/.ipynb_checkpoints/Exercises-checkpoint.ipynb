```python
x = np.array([1, 0, 0])
y = np.array([0, 1, 0])
z = np.array([0, 0, 1])
```

```python
print(np.cross(x, y))
print(np.dot(x, y))
```

```python
a = np.sqrt((np.cross(x, y) * np.cross(x, y)).sum())/(np.sqrt((x*x).sum()) * np.sqrt((y*y).sum()))
np.arcsin(a) * 180/np.pi
```

```python
a = np.sqrt((np.dot(x, y) * np.dot(x, y)).sum())/(np.sqrt((x*x).sum()) * np.sqrt((y*y).sum()))
np.arccos(a) * 180/np.pi
```
