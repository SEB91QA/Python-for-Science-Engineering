```python
import numpy as np


def sphere_area(d: float):
    return np.pi * d ** 2


def sphere_volume(d: float):
    return (np.pi * d ** 3)/6


diameter = []
area = []
volume = []

for i in np.arange(1, 11):
    a = sphere_area(i)
    v = sphere_volume(i)
    diameter.append(i)
    area.append(a)
    volume.append(v)

#np.set_printoptions(formatter={'float': lambda x: format(x, '6.2f')})

diameter = np.array(diameter, dtype=int)
1
area = np.round(np.array(area, dtype=float), decimals=2)
volume = np.round(np.array(volume, dtype=float), decimals=2)

print('diameter' + ' ' + 'area' + ' ' + 'volume')
for i in list(zip(diameter, area, volume)):
    print(i)

print('\ndiameter' + ' ' + 'area' + '      ' + 'volume')
for i in np.arange(0, len(diameter)):
    print(f"{diameter[i]}" + '        ' + f"{area[i]}" + '      ' + f"{volume[i]}")
```

```python
def price(a, f, cpf=1000, add=12000):
    if f <= 2:
        return a * cpf
    else:
        return a * cpf + add*(f-2)


floor46 = price(46, 1200)
floor5 = price(5, 728)

print(
    f"The total cost of an apartment on the 46th floor "
    f"with a total area of 1200 sq.ft is ${floor46:,}"
)

print(
    f"The total cost of an apartment on the 5th floor "
    f"with a total area of 728 sq.ft is ${floor5:,}"
)
```

```python
import numpy as np


def line(x, a, b):
    '''Line function'''
    return a + b*x


def linefit(x, y):
    '''Returns the slope and y-intercept of linear fit to (x, y) data set'''
    xavg = x.mean()
    slope = (y * (x - xavg)).sum() / (x * (x - xavg)).sum()
    yint = y.mean() - slope * xavg
    return yint, slope


def linefitWt(x, y, dy):
    '''Linear regression with Weighting'''
    xbar = ((x/dy**2).sum())/((1/dy*2).sum())
    ybar = ((y/dy**2).sum())/((1/dy*2).sum())

    b = ((x - xbar) * (y/dy**2)).sum()/((x - xbar) * (x/dy**2)).sum()
    a = ybar - b*xbar
    return a, b


def chir(x, y, dy, a, b):
    '''Chi value definition'''
    chi = (((y - a - b*x)/dy)**2).sum()
    return chi/(len(x)-2)


def chir2(f, x, y, dy, *params):  # Revisar las implicaciones teóricas
    model = f(x, *params)
    chi = (((y - model)/dy)**2).sum()
    return chi/(len(y) - 2)


t, y, dy = np.loadtxt('mydata.txt', skiprows=2, unpack=True)
chir(t, y, dy, *linefitWt(t, y, dy))


chir2(line, t, y, dy, *linefitWt(t, y, dy))
```

```python
import math
import numpy as np


def derivA(f, x, h=1.e-9, *params):  # Tuple
    return (f(x + h, *params) - f(x - h, *params)) / (2.*h)


def derivK(f, x, h=1.e-9, **params):  # Dictionary
    return (f(x + h, **params) - f(x - h, **params)) / (2.*h)


def p(x, a, b, c):
    return a + b * x + c * x**2


def V(t, V0, R, C):
    return V0*(1 - np.e ** (-t/(R*C)))


print(derivA(p, -3, 1.e-9, *(2.1, -3.2, 5.8)))
print(derivK(p, -3, 1.e-9, **{'a': 2.1, 'b': -3.2, 'c': 5.8}))
```
