```python
import numpy as np
np.set_printoptions(precision=2)


def sphere_area(d: float):
    return np.pi * d ** 2


def sphere_volume(d: float):
    return (np.pi * d ** 3 )/6


diameter = []
area = []
volume = []

for i in range(1, 11):
    a = sphere_area(i)
    v = sphere_volume(i)
    diameter.append(i)
    area.append(a)
    volume.append(v)

diameter = np.array(diameter, dtype=float)
area = np.array(area)
volume = np.array(volume)

print(diameter)
print(area)
print(volume)
```
